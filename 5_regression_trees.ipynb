```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/real_estate_data.csv")
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
#As you can see, the output displays the number of missing values in each column.
#For instance, the 'CRIM' column has 1 missing value, 'ZN' has 2 missing values, and so on.
#data.isna(): This part utilizes the isna() function, which is specific to Pandas DataFrames.
#It examines each element in the DataFrame and returns a new DataFrame with Boolean values (True or False).
#True indicates a missing value (often represented as NaN - Not a Number), and False indicates a valid value present in the DataFrame.

#.sum(): After isna() creates a DataFrame of True/False values representing missing entries, the .sum() function is applied.
#This function calculates the sum of the elements across the DataFrame.
#Since True translates to 1 and False translates to 0, summing the Booleans essentially counts the number of True values (missing values) in each column.
```

```python
data.dropna(inplace=True)
#The code data.dropna(inplace=True) is a function used in Pandas DataFrames to deal with missing values. Let's break down what each part does:
#data: This represents the Pandas DataFrame you're working with.
#It likely contains columns with data entries, and potentially some missing values (often represented as NaN).
#.dropna(): This part calls the .dropna() method on the DataFrame. This method identifies and removes rows from the DataFrame that contain missing values.
#By default, it removes rows with any missing value (NaN) in any column.
#inplace=True: This is an optional argument within the .dropna() method.
#By setting it to True, you instruct the method to modify the original DataFrame (data) in place.
#This means the rows with missing values are eliminated directly from the DataFrame you assigned to data
data.isna().sum()
```

```python
data.shape
```

```python
x = data.drop(columns=['MEDV'])
y = data['MEDV']
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state = 1)
regression_tree = DecisionTreeRegressor(criterion = 'absolute_error')
regression_tree.fit(x_train, y_train)
regression_tree.score(x_test,y_test)
```

```python
predcition = regression_tree.predict(x_test)
print("$",(predcition - y_test).abs().mean()*1000)
```
